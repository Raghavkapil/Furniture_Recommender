```python
import pandas as pd
from pathlib import Path

CSV = Path("../backend/app/data/intern_data_ikarus.csv")
df = pd.read_csv(CSV)

df["price"] = pd.to_numeric(
    df["price"].astype(str).str.replace(r"[^0-9.\-]", "", regex=True),
    errors="coerce"
)
df.head()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(7,4))
df["price"].dropna().plot(kind="hist", bins=40)
plt.title("Price Distribution")
plt.xlabel("Price"); plt.ylabel("Count")
plt.tight_layout(); plt.show()
```

```python
top_brands = df["brand"].astype(str).value_counts().head(10)
top_cats   = df["categories"].astype(str).value_counts().head(10)
top_brands, top_cats
```

```python
plt.figure(figsize=(7,4))
top_brands.plot(kind="bar"); plt.title("Top Brands")
plt.ylabel("Count"); plt.xticks(rotation=45, ha="right")
plt.tight_layout(); plt.show()

plt.figure(figsize=(7,4))
top_cats.plot(kind="bar"); plt.title("Top Categories")
plt.ylabel("Count"); plt.xticks(rotation=45, ha="right")
plt.tight_layout(); plt.show()
```
